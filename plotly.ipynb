```python
#using plotly express
x = [1, 2, 3, 4]
y = [10, 13, 15, 18]

fig = px.scatter(x,y)
fig.show()
```

```python
fig = px.line(x,y)
fig.show()
```

```python
x = ['A', 'B', 'C']
y = [5, 10, 3]

fig = px.bar(x =x, y=y)
fig.show()
```

```python

x = [5, 10, 3, 4, 5, 5, 6, 6, 10, 10, 3, 11, 11, 12]

fig = px.histogram(x)
fig.show()
```

```python
x = ['A', 'B', 'C']
y = [5, 10, 3, 25, 12, 33, 100]
fig = px.box(x,y)
fig.show()
```

```python
#bubble chart

x = [1, 2, 3, 4]
y = [5, 8, 12, 3]

fig = px.scatter(x, y, size=[30, 10, 50, 200])
fig.show()
```

```python
#3d chart

x = [1, 2, 3, 4]
y = [5, 8, 12, 3]
z = [3, 4, 5, 6]


fig = px.scatter_3d(x=x, y=y, z=z)
fig.show()
```

```python
#comntour plot
x = [1, 2, 3, 4, 3, 23, 80]
y = [5, 8, 12, 3, 12, 35, 60]

fig = px.density_contour(x, y)
fig.show()
```
